```python
address_list_14 = m14['FullAddress'].tolist()
```

```python
len(address_list_14)
```

```python
#temp = address_list[0:5]
longlat_14 = []
```

```python
start_time = time.time()
for i in range(0, len(address_list_14)):
    geolocator = Nominatim()
    location = geolocator.geocode(address_list_14[i])
    if location:
        longlat_14.append((location.latitude, location.longitude))
    else:
        longlat_14.append("")
elapsed_time = time.time() - start_time
```

```python
dictionary = dict(zip(address_list_14, longlat_14 ))
```

```python
m14['LAT_LONG'] = m14['FullAddress'].map(dictionary)
```

```python
m14.to_csv('updated_address.csv')
```
